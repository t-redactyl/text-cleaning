```python
# Open each of the fairytale files
import os
all_files = os.listdir('/Users/jodieburchell/Documents/text-mining/fairytales-corpus-data/')
txt_files = list(filter(lambda x: x[-4:] == '.txt', all_files))
```

```python
# Read all of the books into a list
all_books = []
for t in txt_files:
    with open(t) as f:
        all_books += [f.readlines()]
        
# Strip all of the newline characters and get rid of blank strings
all_books = [[line.strip() for line in book] for book in all_books]
all_books = [list(filter(None, book)) for book in all_books]
```
